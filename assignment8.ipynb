```python
# sns.set(style="whitegrid")

# Create the distribution plot with hue for survival
# plt.figure(figsize=(10, 6))

sns.histplot(df[df['Survived'] == 0]['Age'], bins=10, color='red', label='Not Survived')
sns.histplot(df[df['Survived'] == 1]['Age'], bins=10, color='blue', label='Survived')

plt.title('Distribution of Age Colored by Survival')
plt.xlabel('Age')
# plt.legend()
plt.show()
```

```python
age_group_counts = df.groupby('Age')['Survived'].agg(['count', 'sum'])
age_group_counts['survival_ratio'] = age_group_counts['sum'] / age_group_counts['count'] * 100

# Plot the ratio of survivors upon total count of each age group
plt.figure(figsize=(12, 6))
sns.barplot(x=age_group_counts.index, y='survival_ratio', data=age_group_counts, color='skyblue')
plt.title('Ratio of Survivors Upon Total Count by Age Group')
plt.xlabel('Age')
plt.ylabel('Survival Ratio (%)')
plt.xticks(rotation=90)  # Rotate x-labels for better readability
plt.tight_layout()
plt.show()
```

```python
# percentage of survivers according to age group
df['Age_bin'] = pd.cut(df['Age'], bins=range(0, 90, 10), right=False)

# Calculate the ratio of survivors to total count within each age bin
age_bin_counts = df.groupby('Age_bin')['Survived'].agg(['count', 'sum'])
print(age_bin_counts)
age_bin_counts['survival_ratio'] = age_bin_counts['sum'] / age_bin_counts['count'] * 100
print(age_bin_counts)

# Plot the ratio of survivors upon total count of each age bin
plt.figure(figsize=(10, 6))
sns.barplot(x=age_bin_counts.index.astype(str), y='survival_ratio', data=age_bin_counts, color='skyblue')
plt.title('Ratio of Survivors Upon Total Count by Age Group (Bins of 10)')
plt.xlabel('Age Group')
plt.ylabel('Survival Ratio (%)')
# plt.xticks(rotation=45)  # Rotate x-labels for better readability
plt.tight_layout()
plt.show()
```

```python
pclass_counts = df.groupby('Pclass')['Survived'].agg(['count', 'sum'])
pclass_counts['survival_ratio'] = pclass_counts['sum'] / pclass_counts['count'] * 100

# Plot the ratio of survivors upon total count of each passenger class
plt.figure(figsize=(8, 6))
sns.barplot(x=pclass_counts.index, y='survival_ratio', data=pclass_counts, palette='pastel')
plt.title('Ratio of Survivors Upon Total Count by Passenger Class')
plt.xlabel('Passenger Class')
plt.ylabel('Survival Ratio (%)')
plt.tight_layout()
plt.show()
```

```python
pclass_counts = df.groupby('Sex')['Survived'].agg(['count', 'sum'])
pclass_counts['survival_ratio'] = pclass_counts['sum'] / pclass_counts['count'] * 100

# Plot the ratio of survivors upon total count of each passenger class
plt.figure(figsize=(8, 6))
sns.barplot(x=pclass_counts.index, y='survival_ratio', data=pclass_counts, palette='pastel')
plt.title('Ratio of Survivors Upon Total Count by Passenger Class')
plt.xlabel('Passenger Class')
plt.ylabel('Survival Ratio (%)')
plt.tight_layout()
plt.show()
```
