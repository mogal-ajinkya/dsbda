Use the inbuilt dataset 'titanic' as used in the above problem. Plot a box plot for
distribution of age with respect to each gender along with the information about whether
they survived or not. (Column names : 'sex' and 'age')
Write observations on the inference from the above statistics.

```python
import pandas as pd
df = pd.read_csv('/content/drive/MyDrive/TE/Temp/Dsdba/Titanic-Dataset.csv')
df.head()
```

```python
df = df.drop("Cabin", axis=1)
```

```python
def fun1(value):
    if (value == "male"):
        return 1
    else:
        return 0
def fun2(value):
    if (value == 'S'):
        return 0
    elif (value == 'C'):
        return 1
    elif (value == 'Q'):
        return 2
    else:
        return 0
df["Sex"] = df["Sex"].apply(fun1)
df["Embarked"] = df["Embarked"].apply(fun2)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the Titanic dataset
titanic = sns.load_dataset('titanic')

# Create a box plot
sns.boxplot(x='sex', y='age', hue='survived', data=titanic)

# Set labels and title
plt.xlabel('Gender')
plt.ylabel('Age')
plt.title('Distribution of Age with Respect to Gender and Survival')

# Show the plot
plt.show()
```
